```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from skimage.data import camera
```

```python
image = camera()
plt.imshow(image, cmap = 'gray')
```

```python
rank =  np.linalg.matrix_rank(image)
SVD  =  np.linalg.svd(image)
U = SVD[0]
sigma = SVD[1] # np.diag(sigma)
VT = SVD[2]
```

```python
def Ak(k, U, sigma, VT):
    sum = 0.0
    for i in range(0,k+1):
        ui = U[:,i].reshape(-1,1)
        vti = VT[i,:].reshape(-1,1)
        sigmai = sigma[i]
        sum+= sigmai*np.dot(ui, vti.T)     
    return sum
    
```

```python
images_DR = []
for i in range(0,rank):
    images_DR.append(Ak(i, U, sigma, VT))
```

```python
plt.imshow(images_DR[15], cmap = 'gray')
```
